```r
df<- df %>%
  rename(y = A58) %>%
  mutate(y = ifelse(y == 0, "NotSpam", "Spam")) %>%
  mutate(y = factor(y))
```

```r
library(forcats)
```

```r
df%>%pull(y)%>%fct_count()
```

```r
df %<>% mutate(ID = row_number())

train <- df %>% slice_sample(prop = 0.70)
test  <- df %>% anti_join(train, by = "ID")

train %<>% select(-ID)
test  %<>% select(-ID)
```
