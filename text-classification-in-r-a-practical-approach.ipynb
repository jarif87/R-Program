# 🔗 URL Removal from Tweet

* A custom function `remove_urls()` is defined to clean tweets by removing URLs.
* It uses `gsub("http\\S+", "", tweet)`:
  * `"http\\S+"` is a regular expression matching any URL that starts with "http" followed by non-space characters.
  * `gsub()` replaces all matched URLs with an empty string.
* This helps clean the tweet text for better analysis in NLP tasks (e.g., sentiment analysis or topic modeling).


```r
remove_urls<-function(tweet){
    return(gsub("http\\S+","",tweet))
}
remove_urls(tweet)
```

# 🔁 Removing Retweet Headers with Usernames

* A custom function `remove_usernamewithrt()` is created to clean retweet text.
* It uses the pattern: `^RT @[a-z,A-Z]*[0-9]*[a-z,A-Z]*[0-9]*: `
  * `^RT` ensures the match starts at the beginning of the tweet.
  * `@[a-z,A-Z]*[0-9]*[a-z,A-Z]*[0-9]*` captures a typical Twitter username.
  * The colon `:` and space after the username are also removed.
* `gsub()` replaces the matched retweet header with an empty string.
* This is useful for:
  * Removing attribution in retweets
  * Focusing on the actual message content


```r
remove_usernamewithrt<-function(tweet)
    {
    return(gsub("^RT @[a-z,A-Z]*[0-9]*[a-z,A-Z]*[0-9]*: ","", tweet))
    }

remove_usernamewithrt(tweet)
```

# 👤 Removing User Mentions

* A custom function `remove_username()` is defined to clean tweets by removing user mentions.
* It uses the regular expression: `@[a-z,A-Z]*[0-9]*[a-z,A-Z]*[0-9]*`
  * Matches typical Twitter usernames starting with "@" followed by letters and/or numbers.
  * Does **not** include special characters like underscores or dots.
* `gsub()` replaces all user mentions in the tweet with an empty string.
* Useful for anonymizing data or reducing noise in NLP preprocessing.


```r
remove_username<-function(tweet)
    {
    return(gsub("@[a-z,A-Z]*[0-9]*[a-z,A-Z]*[0-9]*", "", tweet))
    }

remove_username(tweet)
```

# 🏷️ Hashtag Symbol Removal

* A custom function `remove_hashtag()` is defined to clean tweets by removing the `#` symbol.
* It uses `gsub("#", "", tweet)`:
  * Replaces all `#` symbols with an empty string.
  * Keeps the associated text (e.g., `#Headache` becomes `Headache`).
* Useful for:
  * Keeping keyword content while removing unnecessary symbols
  * Preparing text for tokenization or sentiment analysis


```r
remove_hashtag<-function(tweet)
    {
    return(gsub("#","",tweet))

}

remove_hashtag(tweet)
```

# 🧹 Tweet Preprocessing Pipeline

* A new column `processed_tweet` is created in the dataframe to store cleaned tweets.
* Step-by-step cleaning is applied using `apply()` over the tweet text:
  
  * **URL Removal**: Strips out all hyperlinks using `remove_urls()`.
  * **Retweet Username Removal**: Eliminates retweet headers (e.g., `RT @user:`) using `remove_usernamewithrt()`.
  * **Username Removal**: Removes all remaining user mentions like `@someone` using `remove_username()`.
  * **Hashtag Symbol Removal**: Deletes only the `#` symbol while keeping the keyword (e.g., `#pain` → `pain`) using `remove_hashtag()`.

* This step-by-step cleaning prepares the text for further NLP tasks such as tokenization, sentiment analysis, or topic modeling.


```r
df$processed_tweet <- apply(df['tweet'], 2, remove_urls) 
df$processed_tweet <- apply(df['processed_tweet'],2, remove_usernamewithrt) 
df$processed_tweet <- apply(df['processed_tweet'],2, remove_username)
df$processed_tweet <- apply(df['processed_tweet'],2, remove_hashtag)
```

# ⚖️ Word Frequency Comparison Between Annotations

* Words are grouped by `annotation` and their proportions within each group are calculated.
* `pivot_wider()` spreads the data into separate columns for each annotation value (0 = No, 1 = Yes).
* Column `0` is renamed to `No`, and the data is reshaped again for plotting.
* A log-log scatter plot is created using `ggplot()`:
  * `x = proportion` for "Yes", `y = No` for "No"
  * `geom_abline()` adds a diagonal reference line (equal proportion)
  * `geom_jitter()` adds small random variation for better point separation
  * `geom_text()` labels some points with their corresponding word
  * `scale_x_log10()` and `scale_y_log10()` apply log transformation with percentage formatting
  * `scale_color_gradient()` colors points based on how far they deviate from equal proportion
  * `theme(legend.position = "none")` removes the legend for cleaner layout

* **Purpose**:
  * To visualize and compare how frequently each word is used in tweets with `Yes` vs `No` annotations.
  * Words farther from the diagonal line are disproportionately used in one group over the other.


```r
frequency <- text_df %>% count(annotation, word) %>% group_by(annotation) %>%mutate(proportion = n / sum(n)) %>% select(-n) %>% pivot_wider(names_from = annotation, values_from = proportion) %>%
  rename(No = `0`) %>%pivot_longer(3,names_to ="Yes",values_to = "proportion")


ggplot(frequency, aes(x = proportion, y = No,color = abs(No - proportion)))+geom_abline(color = "gray40", lty = 2) + geom_jitter(alpha = 0.1, size = 2.5, width = 0.3, height = 0.3) +
  geom_text(aes(label = word), check_overlap = TRUE, vjust = 1.5) +
  scale_x_log10(labels = percent_format()) +
  scale_y_log10(labels = percent_format()) +
  scale_color_gradient(limits = c(0, 0.001),low = "darkslategray4", high = "gray75") +
  theme(legend.position = "none") + labs(y = "No", x = "Yes", title = "Comparing the Word Frequencies of Different Annotation", subtitle = "Words far from the line are words found more in one set of texts than another")
```

# 🧪 Text Preprocessing Recipe for Modeling

* A `recipe` is created to preprocess the `processed_tweet` column for modeling the `outcome`.

* **Steps included**:
  * `step_tokenize(processed_tweet)` – Splits tweets into individual tokens (words).
  * `step_stopwords(processed_tweet)` – Removes common stop words to reduce noise.
  * `step_stem(processed_tweet)` – Applies stemming to group word variations by their root.
  * `step_tokenfilter(processed_tweet, max_tokens = 1e3)` – Keeps only the 1,000 most frequent tokens to reduce dimensionality.
  * `step_tfidf(processed_tweet)` – Converts tokens into TF-IDF scores to represent their importance within each tweet.

* This preprocessing pipeline ensures clean, consistent, and feature-rich input for machine learning models.


```r
tweet_rec <-recipe(outcome ~ processed_tweet, data = df_train) %>%
  step_tokenize(processed_tweet) %>% # tokenization
  step_stopwords(processed_tweet)%>% # stopwords removal
  step_stem(processed_tweet) %>% #stem
  step_tokenfilter(processed_tweet, max_tokens = 1e3) %>% # select tokens
  step_tfidf(processed_tweet)
```

# 🔁 Upsampling Recipe for Imbalanced Data

* A `recipe` named `tweet_upsample_rec` is created to preprocess tweet text and address class imbalance.

* **Included Steps**:
  * `step_tokenize(processed_tweet)` – Tokenizes the tweet text into individual words.
  * `step_stopwords(processed_tweet)` – Removes stop words to reduce noise.
  * `step_stem(processed_tweet)` – Applies stemming to normalize word forms.
  * `step_tokenfilter(processed_tweet, max_tokens = 1e3)` – Retains the top 1,000 most frequent tokens.
  * `step_tfidf(processed_tweet)` – Transforms tokens into TF-IDF features.
  * `step_upsample(outcome, over_ratio = 1)` – Balances the classes by duplicating minority class samples until both classes have equal frequency.

* **Purpose**:
  * To improve model fairness and predictive performance on underrepresented classes by balancing the outcome variable during training.


```r
tweet_upsample_rec <- recipe(outcome ~ processed_tweet,data = df_train) %>%
  step_tokenize(processed_tweet) %>%
  step_stopwords(processed_tweet)%>%
  step_stem(processed_tweet) %>%
  step_tokenfilter(processed_tweet, max_tokens = 1e3) %>%
  step_tfidf(processed_tweet) %>%
  step_upsample(outcome, over_ratio = 1)
```

```r
tweet_upsample_rec
```

# 🌟 Variable Importance Analysis

* `extract_fit_parsnip(final_wf$.workflow[[1]])`:
  * Extracts the fitted parsnip model object from the final workflow.
  
* `vip::vi()`:
  * Computes variable importance scores for model features.
  * Returns a tibble with variables, importance values, and sign (positive/negative influence).

* Filtering and arranging:
  * Displays the top 10 positively influential features (`Sign == "POS"`) by descending importance.

* Data cleaning and transformation:
  * `Sign` is relabeled for clarity:
    * "POS" → "More about Plausible Covid-19 Cases"
    * "NEG" → "Less about Plausible Covid-19 Cases"
  * Importance values are converted to absolute values for ranking.
  * Variable names are cleaned by removing text prefixes.

* Visualization with `ggplot2`:
  * Features are grouped by `Sign` (positive/negative influence).
  * The top 20 features per group are plotted as horizontal bars.
  * Facets separate positive and negative features for comparison.
  * The plot highlights which features most influence prediction of plausible COVID-19 related tweets.

* This analysis provides interpretability into which words (features) drive the model’s decisions.


```r
library(vip)
tidymodels_prefer()
final_imp <- extract_fit_parsnip(final_wf$.workflow[[1]]) %>%
  vip::vi()

final_imp%>%filter(Sign == "POS") %>% arrange(desc(Importance)) %>% head(10)

final_imp %>%
  mutate(
    Sign = case_when(Sign == "POS" ~ "More about Plausible Covid-19 Cases",
                     Sign == "NEG" ~ "Less about Plausible Covid-19 Cases"),
    Importance = abs(Importance),
    Variable = str_remove_all(Variable, "tfidf_processed_tweet_"),
    Variable = str_remove_all(Variable, "textfeature_narrative_copy_")
  ) %>%
  group_by(Sign) %>%
  top_n(20, Importance) %>%
  ungroup %>%
  ggplot(aes(x = Importance,
             y = fct_reorder(Variable, Importance),
             fill = Sign)) +
  geom_col(show.legend = FALSE) +
  scale_x_continuous(expand = c(0, 0)) +
  facet_wrap(~Sign, scales = "free") +
  labs(
    y = NULL,
    title = "Variable importance for predicting Plausible COVID-19 Cases",
    subtitle = paste0("These features are the most important in predicting\n",
                      "whether a tweet is about Plausible COVID-19 Cases or not")
  )
```
