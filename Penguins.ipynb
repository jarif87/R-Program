```r
options(repr.plot.width = 20, repr.plot.height = 10)
scatterplotMatrix(~ bill_length_mm +bill_depth_mm +flipper_length_mm +body_mass_g | y,
  data = df,
  regLine = FALSE,
  smooth  = FALSE,
  pch = c(2, 6))
```

```r
# calculate cluster
hc<-df%>%dist%>%hclust
```

```r
hc$labels <- y  # Set the class labels
```

```r
# plot the dendrogram
hc%>%plot(hang=-1,cex=0.6)
```

```r
install.packages("factoextra")
library(factoextra)
```

```r
# Elbow method

df%>%fviz_nbclust(FUN=hcut,method="wss")+geom_vline(xintercept = 3,color="red",linetype="dotted")
```

```r

# Silhouette method

df%>%fviz_nbclust(FUN=hcut,method="silhouette")
```

```r
install.packages("cluster")   # if not installed
library(cluster)
```

```r
# Gap method: This compares the total intracluster variation
# for values of k with their expected values from null
# distributions (from Monte Carlo simulations)

df%>%clusGap(FUN=hcut,K.max=10,B=100)%>%fviz_gap_stat()
```

```r
# plot dendogram
hc%>%plot(hang=-1,cex=0.8)
```

```r
# Create hierarchical clustering object
hc <- hclust(dist(df), method = "ward.D2")

# Plot the dendrogram
plot(hc, hang = -1, main = "Dendrogram of Clustering")

# Now add the rectangles
rect.hclust(hc, k = 3, border = 2:4)
```

```r
y_hc=hc%>%cutree(3)
```

```r
fviz_cluster(list(cluster=y_hc,data=df),geom=c("point")+geom_text(vjust=1.5,aes(color=as.factor(y_hc),
label=hc$labels)))
```
