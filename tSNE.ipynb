```r
library(magrittr)
library(dplyr)
library(Rtsne)
```

```r
df<-import("/content/optdigits.csv")
```

```r
head(df,5)
```

```r
df<-df%>%rename(y=X0.26)
```

```r
df%>%select(-y)%>%Rtsne(perplexity=1,verbose=TRUE,max_iter=500)%>%.$Y%>%plot(
    col=df$y,pch=19
)
```

```r
df %>%
  select(-y) %>%
  Rtsne(
    perplexity = 2,  # Value to vary
    verbose = TRUE,
    max_iter = 500
  ) %>%
  .$Y %>%
  plot(
    col = df$y,
    pch = 19
  )
```

```r
df %>%
  select(-y) %>%
  Rtsne(
    perplexity = 50,  # Value to vary
    verbose = TRUE,
    max_iter = 500
  ) %>%
  .$Y %>%
  plot(
    col = df$y,
    pch = 19
  )
```
